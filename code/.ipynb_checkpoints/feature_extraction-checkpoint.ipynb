```python
import json
import pickle
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

with open('../data/training_data.obj', 'rb') as training_file:
    training_data = pickle.load(training_file)
with open('../data/sentiment_words.json') as sent_words:
    sentiment_words = json.load(sent_words)
    
```

#### Get rough sentiment of tweet

```python
tweets = []
for tweet in training_data['tweets']:
    score = 0
    for word in tweet[1].split():
        root_word = stemmer.stem(word)
        if root_word in sentiment_words:
            score += sentiment_words[root_word]
    tweets.append([tweet[6], score])
tweets
```
